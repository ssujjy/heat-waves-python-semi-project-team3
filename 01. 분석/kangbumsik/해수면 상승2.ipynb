```python
file_path = "../file/PIOMAS.thick.daily.1979.2024.Current.v2.1.csv"
output_file_path = "../file/PIOMAS.thick.daily.1979.2024.Current.v2.1_modified.csv"

# 파일을 읽기 모드로 열기
with open(file_path, "r") as file:
    # 파일의 내용 읽기
    file_content = file.read()

# 공백을 쉼표로 대체하기
modified_content = ""
for line in file_content.splitlines():
    modified_line = ','.join(line.split())
    modified_content += modified_line + '\n'


# 수정된 내용을 새로운 파일에 쓰기
with open(output_file_path, "w") as output_file:
    output_file.write(modified_content)
```
