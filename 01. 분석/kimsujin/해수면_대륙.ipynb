```python
import pandas as pd
```

```python
# Data 불러오기

# data1990 = pd.read_csv("Data/배출원별_온실가스_배출량_1991_2000.csv", encoding='UTF-8',index_col=0)
# data2000 = pd.read_csv("Data/배출원별_온실가스_배출량_2001_2010.csv", encoding='UTF-8',index_col=0)
data2010 = pd.read_csv("Data/배출원별_온실가스_배출량_2011_2021.csv", encoding='UTF-8',index_col=0)
```

```python
# df1990 = pd.DataFrame(data1990)
# df2000 = pd.DataFrame(data2000)
df2010 = pd.DataFrame(data2010)
```

---
년도별로 가져온 데이터 나누기

```python
# 첫 5개 열 선택
df_2011 = df2010.iloc[:, 0:5]

# 첫 번째 row의 값을 colnames로 설정
df_2011.columns = df_2011.iloc[0]

# colnames로 해준 row값 제거
df_2011 = df_2011[1:]
```

---
대륙별로 구분하기 위해 대륙이름과 대륙이름이 있는 인덱스 번호 가져오기

```python
# 모든 값이 NaN인 행의 인덱스 찾기.
df_2011[(df_2011.isna().all(axis=1))]    
```

```python
# 모든 값이 NaN인 행의 인덱스 찾아서 대륙명(continent_name)에 넣기.
continent_name = df_2011.index[df_2011.isna().all(axis=1)].tolist()
continent_name
```

```python
# 인덱스로 되어있는 국가를 컬럼으로 넣고 인덱스를 숫자로 변경.

df_2011.reset_index(inplace=True)
# df_2011.reset_index(level=None, drop=False, inplace=False, col_level=0, col_fill='')

# 국가가 있는 행만 빼고 나머지행을 임시로 할당.
temp_2011 = df_2011.iloc[:,1:]

temp_2011
```

```python
continent_idx = temp_2011.index[temp_2011.isna().all(axis=1)].tolist()
continent_idx
```

```python
continent = pd.DataFrame(continent_idx, continent_name) 
continent
```

---
# 대륙별로 나눠서 데이터 프레임 생성

-> 함수를 만들어서 하려고하다가 데이터를 하나씩 확인하면서 작업하기로 함.

```python
# print(df_2011.iloc[0:11,:])
df_2011_asia = df_2011.iloc[1:11,:]
# df_2011_asia['대륙'] = '아시아'
df_2011_asia.insert(0,'대륙','아시아')
df_2011_asia
```

```python
# print(df_2011.iloc[11:15,:])
df_2011_north_afria = df_2011.iloc[12:15,:]
df_2011_north_afria.insert(0,'대륙','북아메리카')
df_2011_north_afria
```

```python
print(df_2011.iloc[15:22,:])
df_2011_south_afria = df_2011.iloc[16:22,:]
df_2011_south_afria.insert(0,'대륙','남아메리카')
df_2011_south_afria
```

```python
# print(df_2011.iloc[22:58,:])
df_2011_Europe = df_2011.iloc[23:58,:]
df_2011_Europe.insert(0,'대륙','유럽')
df_2011_Europe
```

```python
print(df_2011.iloc[58:60,:])
df_2011_afria = df_2011.iloc[59:60,:]
df_2011_afria.insert(0,'대륙','아프리카')
df_2011_afria
```

```python
print(df_2011.iloc[60:,:])
df_2011_oceania = df_2011.iloc[61:,:]
df_2011_oceania.insert(0,'대륙','오세아니아')
df_2011_oceania
```

---
### 대륙별로 나눈 데이터를 하나로 합하기

```python
df_2011_earth = pd.concat([df_2011_asia,df_2011_north_afria,df_2011_south_afria,df_2011_Europe,df_2011_afria,df_2011_oceania],ignore_index=True)
df_2011_earth
```

```python
df_2011_earth.to_csv("Data/배출원별_온실가스_배출량_2011.csv", index=False)
```
